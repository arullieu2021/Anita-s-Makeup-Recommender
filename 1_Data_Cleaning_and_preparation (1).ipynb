#Data Cleaning

Drop ["created_at", "updated_at", "product_api_url", "api_featured_image"]

```python
cols_to_drop = ["created_at", "updated_at", "product_api_url", "api_featured_image"]

clean_products_df = products_df.drop(columns=cols_to_drop, errors="ignore")
clean_products_df.head()
```

Drop empty colours rows


```python
empty_colors = clean_products_df[clean_products_df["product_colors"].astype(str).str.strip() == "[]"]

print(len(empty_colors), "rows with empty product_colors")
empty_colors.head()
```

```python
# Remove rows where product_colors is "[]" or NaN
clean_products_df = clean_products_df[
    ~(clean_products_df["product_colors"].astype(str).str.strip() == "[]")
    & ~clean_products_df["product_colors"].isna()
]

print(f"Remaining products: {len(clean_products_df)}")
clean_products_df.head()
```

See remaining products

```python
import ast

rows = []
for _, row in clean_products_df.iterrows():
    colors = row.get("product_colors")
    try:
        colors_list = ast.literal_eval(colors)
    except (ValueError, SyntaxError):
        continue

    for c in colors_list:
        rows.append({
            "id": row["id"],
            "brand": row["brand"],
            "name": row["name"],
            "product_type": row["product_type"],
            "hex": c["hex_value"].lstrip("#"),
            "shade_name": c["colour_name"],
        })

catalog_df = pd.DataFrame(rows)
print(f"Total shades in catalog: {len(catalog_df)}")
catalog_df.head()
```

#Show NaN Values

```python
# 1. Summary table: count and percentage of NaN per column
nan_summary = pd.DataFrame({
    "nan_count":   clean_products_df.isna().sum(),
    "nan_percent": (clean_products_df.isna().mean() * 100).round(2)
})
nan_summary = nan_summary[nan_summary["nan_count"] > 0].sort_values("nan_percent", ascending=False)
print("=== Columns with NaN values ===")
print(nan_summary.to_string())

# 2. Rows where key business columns are NaN
important_cols = ["brand", "name", "price", "rating", "product_type", "category", "description"]
print("\n=== NaN counts in important columns ===")
for col in important_cols:
    n = clean_products_df[col].isna().sum()
    if n:
        print(f"  {col}: {n} NaN rows")
        print(clean_products_df[clean_products_df[col].isna()][["id", "brand", "name", "price", "product_type"]].head(3).to_string(index=False))
        print()

# 3. Rows with ANY NaN across all columns
rows_with_any_nan = clean_products_df[clean_products_df.isna().any(axis=1)]
print(f"=== Rows with at least one NaN: {len(rows_with_any_nan)} / {len(clean_products_df)} ===")
print(rows_with_any_nan[important_cols].head(10).to_string())
```

Drop Nan Values and fill more columns

```python
import pandas as pd
import re

# ── 1. Drop NaN brand (8 rows) ─────────────────────────────────────────────
clean_products_df = clean_products_df.dropna(subset=["brand"])

# ── 2. Fill currency / price_sign with defaults ────────────────────────────
clean_products_df["price_sign"] = clean_products_df["price_sign"].fillna("$")
clean_products_df["currency"]   = clean_products_df["currency"].fillna("USD")

# ── 3. Remove nail polish ──────────────────────────────────────────────────
nail_polish_mask = (
    clean_products_df["product_type"].str.contains("nail", case=False, na=False) |
    clean_products_df["category"].str.contains("nail", case=False, na=False) |
    clean_products_df["name"].str.contains("nail polish", case=False, na=False)
)
clean_products_df = clean_products_df[~nail_polish_mask]

# ── 4. Strip HTML from description, fill remaining NaN with "" ─────────────
def strip_html(text):
    if pd.isna(text):
        return ""
    return re.sub(r"<[^>]+>", "", str(text)).strip()

clean_products_df["description"] = clean_products_df["description"].apply(strip_html)

# ── 5. Drop NaN price (10 rows) ────────────────────────────────────────────
clean_products_df = clean_products_df.dropna(subset=["price"])

# ── 6. Fill category with product_type where missing ──────────────────────
clean_products_df["category"] = clean_products_df["category"].fillna(
    clean_products_df["product_type"]
)

# ── 7. Drop rating (67% NaN — not reliable) ────────────────────────────────
clean_products_df = clean_products_df.drop(columns=["rating"])

# ── 8. Summary ─────────────────────────────────────────────────────────────
print(f"Final dataset: {len(clean_products_df)} rows")
print(f"Columns: {list(clean_products_df.columns)}\n")
print("=== Remaining NaN ===")
remaining = clean_products_df.isna().sum()
remaining = remaining[remaining > 0]
print(remaining if not remaining.empty else "None — dataset is clean!")
```

#Reassign the ID column

```python
clean_products_df = clean_products_df.reset_index(drop=True)
clean_products_df["id"] = clean_products_df.index + 1

print(f"New id range: {clean_products_df['id'].min()} → {clean_products_df['id'].max()}")
clean_products_df[["id", "brand", "name"]].head()
```

#Add a new column : undertone

```python
!pip install colormath -q
```

```python
import ast
from colormath.color_objects import sRGBColor, LabColor
from colormath.color_conversions import convert_color
import numpy as np

def hex_to_hue_angle(hex_str):
    hex_str = hex_str.strip().lstrip("#")
    r, g, b = int(hex_str[0:2],16)/255, int(hex_str[2:4],16)/255, int(hex_str[4:6],16)/255
    rgb = sRGBColor(r, g, b)
    lab = convert_color(rgb, LabColor)
    return np.degrees(np.arctan2(lab.lab_b, lab.lab_a)) % 360

def assign_undertone(product_colors_str):
    try:
        colors_list = ast.literal_eval(product_colors_str)
    except (ValueError, SyntaxError):
        return "neutral"

    angles = []
    for c in colors_list:
        try:
            angles.append(hex_to_hue_angle(c["hex_value"]))
        except Exception:
            continue

    if not angles:
        return "neutral"

    avg_angle = np.mean(angles)

    if avg_angle > 60:
        return "warm"
    elif avg_angle < 60:
        return "cool"
    else:
        return "neutral"

# ── Add undertone column ───────────────────────────────────────────────────
clean_products_df["undertone"] = clean_products_df["product_colors"].apply(assign_undertone)

print(clean_products_df["undertone"].value_counts())
clean_products_df[["brand", "name", "product_type", "undertone"]].head(10)
```
